```python
import pandas as pd

a = {
    'EmployeeID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Department': ['HR', 'IT', 'IT', 'HR']
}
df1= pd.DataFrame(a)

b = {
    'EmployeeID': [3, 4, 5, 6],
    'Salary': [65000, 52000, 70000, 71000],
    'JoiningDate': ['2020-01-15', '2019-03-23', '2021-06-12', '2018-09-17']
}
df2 = pd.DataFrame(b)
```

```python
#inner Merge

#Merge dataFrame on ID
inner = pd.merge(df1,df2, on ='EmployeeID')

inner

```

```python
#outer Merge

outer = pd.merge(df1,df2, on ='EmployeeID', how='outer')

outer
```

```python
#left Merge

Left=pd.merge(df1,df2, on ='EmployeeID', how= 'left')
Left
```

```python
#left Merge

Right=pd.merge(df1,df2, on ='EmployeeID', how= 'right')
Right
```
